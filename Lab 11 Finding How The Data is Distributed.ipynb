### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# This script analyzes and visualizes the comparison between languages developers
# have worked with and languages they want to work with.

def get_language_counts(series):
    """
    Parses a pandas Series of semi-colon separated strings (languages)
    and returns a Counter object with the frequency of each language.
    """
    # Drop missing values to avoid errors
    series_nonull = series.dropna()
    # Use a Counter for efficient counting
    lang_counter = Counter()
    # Split each string of languages and update the counter
    for lang_string in series_nonull:
        lang_counter.update(lang_string.split(';'))
    return lang_counter

try:
    # Attempt to load the dataset.
    df = pd.read_csv(data_url)

    # --- Data Processing ---
    # Get the frequency counts for both columns using the helper function.
    worked_with_counts = get_language_counts(df['LanguageHaveWorkedWith'])
    want_to_work_counts = get_language_counts(df['LanguageWantToWorkWith'])

    # Convert the Counter objects to pandas Series for easier manipulation.
    worked_series = pd.Series(worked_with_counts)
    want_series = pd.Series(want_to_work_counts)

    # Combine the two series into a single DataFrame.
    # The index will be the language name.
    lang_df = pd.DataFrame({
        'HaveWorkedWith': worked_series,
        'WantToWorkWith': want_series
    })

    # --- Prepare for Plotting ---
    # To make the chart readable, we'll focus on the most popular languages.
    # We sort by 'HaveWorkedWith' and take the top 15.
    top_n = 15
    top_languages_df = lang_df.nlargest(top_n, 'HaveWorkedWith')

    # --- Visualization ---
    sns.set_style("whitegrid")
    ax = top_languages_df.plot(
        kind='bar',
        figsize=(15, 8),
        width=0.8, # Width of the bars
        color=sns.color_palette("viridis", 2) # Use a nice color palette
    )

    # Set titles and labels for clarity
    plt.title(f'Top {top_n} Languages: Current Use vs. Future Desire', fontsize=18, pad=20)
    plt.xlabel('Programming Languages', fontsize=12)
    plt.ylabel('Number of Respondents', fontsize=12)

    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add a legend to distinguish the bars
    plt.legend(title='Category', labels=['Have Worked With', 'Want to Work With'])

    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()

    print("Displaying grouped bar chart for language comparison...")
    plt.show()


except FileNotFoundError:
    print("\nError: The file 'survey_results_public.csv' was not found.")
    print("Please ensure the dataset file is in the correct directory and try again.")
except KeyError as e:
    print(f"\nError: Column not found in the dataset: {e}")
    print("Please check the column names in your CSV file.")

```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# This script visualizes the distribution of remote work status
# across different countries from the Stack Overflow Developer Survey.

try:
    # Attempt to load the dataset.
    df = pd.read_csv(data_url)

    # --- Data Preparation ---
    # To make the chart readable, let's first find the top 10 countries
    # with the most respondents.
    top_countries = df['Country'].value_counts().nlargest(10).index

    # Filter the DataFrame to include only data from these top countries.
    df_top_countries = df[df['Country'].isin(top_countries)]

    # Clean the 'RemoteWork' column by dropping rows where it's missing
    # to ensure accurate counts.
    df_cleaned = df_top_countries.dropna(subset=['RemoteWork'])

    # --- Create a Contingency Table ---
    # pandas.crosstab is a great way to create a frequency table of two or more factors.
    # Here, we count the occurrences of each 'RemoteWork' status for each 'Country'.
    contingency_table = pd.crosstab(df_cleaned['Country'], df_cleaned['RemoteWork'])

    # Reorder the columns for a logical flow in the chart (e.g., from remote to in-person)
    # This step is optional but improves readability.
    desired_order = ['Fully remote', 'Hybrid (some remote, some in-person)', 'Full-time office']
    # Find which of the desired columns actually exist in the table
    existing_order = [col for col in desired_order if col in contingency_table.columns]
    contingency_table = contingency_table[existing_order]


    # --- Visualization (Grouped Bar Chart) ---
    sns.set_style("whitegrid")
    ax = contingency_table.plot(
        kind='bar',
        figsize=(16, 9),
        width=0.8,
        color=sns.color_palette("viridis", len(contingency_table.columns))
    )

    # Set titles and labels for clarity
    plt.title('Distribution of Remote Work Status in Top 10 Countries', fontsize=18, pad=20)
    plt.xlabel('Country', fontsize=12)
    plt.ylabel('Number of Respondents', fontsize=12)

    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add a legend to distinguish the bars
    plt.legend(title='Remote Work Status')

    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()

    print("Displaying grouped bar chart for remote work distribution...")
    plt.show()


except FileNotFoundError:
    print("\nError: The file 'survey_results_public.csv' was not found.")
    print("Please ensure the dataset file is in the correct directory and try again.")
except KeyError as e:
    print(f"\nError: Column not found in the dataset: {e}")
    print("Please check the column names in your CSV file.")

```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install pandas matplotlib seaborn scipy
```

```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# This script analyzes the correlation between job satisfaction ('JobSat')
# and years of professional coding experience ('YearsCodePro').

try:
    # Attempt to load the dataset.
    df = pd.read_csv(data_url)

    # --- Data Preparation ---
    # Create a copy to avoid modifying the original DataFrame in memory
    df_corr = df[['JobSat', 'YearsCodePro']].copy()

    # 1. Convert 'YearsCodePro' to a numeric type
    # Handle special string values first
    df_corr['YearsCodePro'] = df_corr['YearsCodePro'].replace({
        'Less than 1 year': 0,
        'More than 50 years': 51
    })
    # Convert the column to a numeric type, coercing errors to NaN
    df_corr['YearsCodePro'] = pd.to_numeric(df_corr['YearsCodePro'], errors='coerce')

    # 2. Convert 'JobSat' to an ordered numerical scale (ordinal encoding)
    satisfaction_mapping = {
        'Very dissatisfied': 1,
        'Slightly dissatisfied': 2,
        'Neither satisfied nor dissatisfied': 3,
        'Slightly satisfied': 4,
        'Very satisfied': 5
    }
    df_corr['JobSat_Num'] = df_corr['JobSat'].map(satisfaction_mapping)

    # 3. Drop rows with missing values in either of the key columns
    df_cleaned = df_corr.dropna(subset=['JobSat_Num', 'YearsCodePro'])

    # --- Correlation Calculation ---
    # Calculate the Spearman rank correlation coefficient and the p-value.
    # Spearman is used because 'JobSat' is ordinal data.
    corr, p_value = spearmanr(df_cleaned['JobSat_Num'], df_cleaned['YearsCodePro'])

    print("--- Correlation Analysis: Job Satisfaction vs. Years of Experience ---")
    print(f"Spearman Correlation Coefficient: {corr:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation of the results
    if p_value < 0.05:
        print("The p-value is statistically significant, suggesting the correlation is not due to random chance.")
        if corr > 0.2:
            print("There is a small to moderate positive correlation.")
        elif corr > 0:
            print("There is a weak positive correlation.")
        elif corr < -0.2:
            print("There is a small to moderate negative correlation.")
        else:
            print("There is a weak negative correlation.")
    else:
        print("The p-value is not statistically significant, so we cannot conclude there is a correlation.")
    print("-" * 60)


    # --- Visualization (Box Plot) ---
    # A box plot is excellent for showing the distribution of a numerical variable
    # across different categories of an ordinal variable.
    plt.figure(figsize=(12, 7))
    sns.boxplot(
        x='JobSat',
        y='YearsCodePro',
        data=df_cleaned,
        order=satisfaction_mapping.keys(), # Ensure correct order on the x-axis
        palette='viridis'
    )

    # Set titles and labels
    plt.title('Job Satisfaction vs. Years of Professional Coding Experience', fontsize=16, pad=20)
    plt.xlabel('Job Satisfaction', fontsize=12)
    plt.ylabel('Years of Professional Coding Experience', fontsize=12)
    plt.xticks(rotation=15, ha='right')

    plt.tight_layout()
    print("\nDisplaying box plot for visualization...")
    plt.show()


except FileNotFoundError:
    print("\nError: The file 'survey_results_public.csv' was not found.")
    print("Please ensure the dataset file is in the correct directory and try again.")
except KeyError as e:
    print(f"\nError: Column not found in the dataset: {e}")
    print("Please check the column names in your CSV file.")

```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# This script analyzes the relationship between employment status ('Employment')
# and education level ('EdLevel') from the Stack Overflow Developer Survey.

try:
    # Attempt to load the dataset.
    df = pd.read_csv(data_url)

    # --- Data Preparation ---
    # To keep the visualization clean, we'll focus on the most common employment statuses.
    top_employment_statuses = df['Employment'].value_counts().nlargest(5).index
    df_filtered = df[df['Employment'].isin(top_employment_statuses)]

    # Also, let's simplify the education level categories for better readability
    # by grouping some of them.
    def simplify_edlevel(level):
        if 'Bachelor’s degree' in str(level):
            return 'Bachelor’s degree'
        if 'Master’s degree' in str(level):
            return 'Master’s degree'
        if 'Some college/university study' in str(level):
            return 'Some college/university study'
        if 'Associate degree' in str(level):
            return 'Associate degree'
        if 'Professional degree' in str(level) or 'Other doctoral' in str(level):
            return 'Doctoral/Professional degree'
        return 'Other' # Group less common categories

    df_filtered['EdLevel_Simple'] = df_filtered['EdLevel'].apply(simplify_edlevel)

    # --- Cross-tabulation ---
    # Create a frequency table of Employment Status vs. Simplified Education Level.
    crosstab_ed = pd.crosstab(df_filtered['Employment'], df_filtered['EdLevel_Simple'])

    # To make comparison easier, we'll calculate the percentage of each education
    # level within each employment category (normalize by row).
    crosstab_pct = crosstab_ed.apply(lambda r: r/r.sum() * 100, axis=1)

    print("--- Crosstab: Employment Status vs. Education Level (%) ---")
    print(crosstab_pct)
    print("-" * 60)


    # --- Visualization (Stacked Bar Plot) ---
    sns.set_style("whitegrid")
    ax = crosstab_pct.plot(
        kind='bar',
        stacked=True,
        figsize=(14, 8),
        colormap='viridis'
    )

    # Set titles and labels for clarity
    plt.title('Education Level Distribution by Employment Status', fontsize=16, pad=20)
    plt.xlabel('Employment Status', fontsize=12)
    plt.ylabel('Percentage of Respondents (%)', fontsize=12)
    plt.xticks(rotation=15, ha='right')

    # Move the legend outside of the plot for better visibility
    plt.legend(title='Education Level', bbox_to_anchor=(1.02, 1), loc='upper left')

    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()

    print("\nDisplaying stacked bar plot for visualization...")
    plt.show()


except FileNotFoundError:
    print("\nError: The file 'survey_results_public.csv' was not found.")
    print("Please ensure the dataset file is in the correct directory and try again.")
except KeyError as e:
    print(f"\nError: Column not found in the dataset: {e}")
    print("Please check the column names in your CSV file.")
```
