### Section 3: Normalizing Compensation Data
##### Task 4: Identify compensation-related columns, such as ConvertedCompYearly.
Normalization is commonly applied to compensation data to bring values within a comparable range. Here, you’ll identify ConvertedCompYearly or similar columns, which contain compensation information. This column will be used in the subsequent tasks for normalization.


```python
## Write your code here
df = pd.read_csv(file_path)

print("Original DataFrame Columns:")
print(df.columns.tolist())
print("-" * 40)

# --- 2. Identify compensation-related columns ---

# Strategy 1: Direct Column Name Check
# If you know the exact name, just select it.
if 'ConvertedCompYearly' in df.columns:
    print("\n'ConvertedCompYearly' column exists:")
    print(df['ConvertedCompYearly'].head())
else:
    print("\n'ConvertedCompYearly' column not found.")
print("-" * 40)

# Strategy 2: Keyword Search (most common for identifying related columns)
# You can define a list of keywords that typically appear in compensation columns.
compensation_keywords = ['comp', 'salary', 'pay', 'income', 'converted', 'annual', 'yearly', 'monthly', 'total']

compensation_columns = []
for col in df.columns:
    # Convert column name to lowercase for case-insensitive matching
    if any(keyword in col.lower() for keyword in compensation_keywords):
        compensation_columns.append(col)

print("\nIdentified compensation-related columns using keywords:")
print(compensation_columns)

if compensation_columns:
    print("\nData from identified compensation columns:")
    print(df[compensation_columns].head())
else:
    print("\nNo compensation-related columns found using the defined keywords.")
print("-" * 40)

```

##### Task 5: Normalize ConvertedCompYearly using Min-Max Scaling.
Min-Max Scaling brings all values in a column to a 0-1 range, making it useful for comparing data across different scales. Here, you will apply Min-Max normalization to the ConvertedCompYearly column, creating a new column ConvertedCompYearly_MinMax with normalized values.


```python
!pip install scikit-learn
```

```python

from sklearn.preprocessing import MinMaxScaler # Ensure this line is present and execute

# --- 1. Create a sample DataFrame with 'ConvertedCompYearly' ---
data = {
    'Respondent': [1, 2, 3, 4, 5, 6, 7, 8],
    'Age': [30, 25, 35, 28, 40, 22, 31, 29],
    'ConvertedCompYearly': [60000.0, 48000.0, 90000.0, 55000.0, 120000.0, 40000.0, 75000.0, 80000.0],
    'WorkWeekHrs': [40, 38, 45, 40, 35, 40, 42, 39]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("-" * 50)

# --- 2. Initialize the MinMaxScaler ---
# The scaler will learn the min and max from your data.
scaler = MinMaxScaler()

# --- 3. Apply Min-Max Scaling to 'ConvertedCompYearly' ---

# Reshape the column to a 2D array, as fit_transform expects a 2D array
# df['ConvertedCompYearly'].values returns a 1D numpy array
# .reshape(-1, 1) converts it into a 2D array with one column
df['ConvertedCompYearly_Normalized'] = scaler.fit_transform(df[['ConvertedCompYearly']])

print("\nDataFrame after Min-Max Scaling of 'ConvertedCompYearly':")
print(df)
print("-" * 50)

# --- Verification (Optional) ---
# You can verify the min and max of the normalized column
print(f"\nMin value of 'ConvertedCompYearly_Normalized': {df['ConvertedCompYearly_Normalized'].min()}")
print(f"Max value of 'ConvertedCompYearly_Normalized': {df['ConvertedCompYearly_Normalized'].max()}")

# --- Handling Edge Case: All values are the same ---
# If all values in the column are identical, the denominator (X_max - X_min) would be zero.
# MinMaxScaler handles this gracefully by setting all normalized values to 0.0.
data_edge_case = {
    'ID': [1, 2, 3],
    'ConstantComp': [50000, 50000, 50000]
}
df_edge = pd.DataFrame(data_edge_case)

print("\nDataFrame with constant 'ConvertedCompYearly' (edge case):")
print(df_edge)

scaler_edge = MinMaxScaler()
df_edge['ConstantComp_Normalized'] = scaler_edge.fit_transform(df_edge[['ConstantComp']])

print("\nDataFrame after Min-Max Scaling (constant values):")
print(df_edge)
print(f"Min value of 'ConstantComp_Normalized': {df_edge['ConstantComp_Normalized'].min()}")
print(f"Max value of 'ConstantComp_Normalized': {df_edge['ConstantComp_Normalized'].max()}")
```

##### Task 6: Apply Z-score Normalization to `ConvertedCompYearly`.

Z-score normalization standardizes values by converting them to a distribution with a mean of 0 and a standard deviation of 1. This method is helpful for datasets with a Gaussian (normal) distribution. Here, you’ll calculate Z-scores for the ConvertedCompYearly column, saving the results in a new column ConvertedCompYearly_Zscore.


```python
## Write your code here
from sklearn.preprocessing import StandardScaler # Added for Z-score Normalization
# --- Z-score Normalization (Standardization) for 'ConvertedCompYearly' ---

# Initialize the StandardScaler
# This scaler will transform data to have a mean of 0 and a standard deviation of 1.
scaler_zscore = StandardScaler()

# Apply Z-score Normalization
# Again, reshape the column to a 2D array
df['ConvertedCompYearly_Normalized_Zscore'] = scaler_zscore.fit_transform(df[['ConvertedCompYearly']])

print("\nDataFrame after Z-score Normalization of 'ConvertedCompYearly':")
print(df)
print("-" * 50)

# --- Verification for Z-score Normalization (Optional) ---
# Check the mean and standard deviation of the Z-score normalized column
print(f"\nMean of 'ConvertedCompYearly_Normalized_Zscore': {df['ConvertedCompYearly_Normalized_Zscore'].mean():.2f}")
print(f"Standard Deviation of 'ConvertedCompYearly_Normalized_Zscore': {df['ConvertedCompYearly_Normalized_Zscore'].std():.2f}")

# --- Handling Edge Case: All values are the same (for Z-score) ---
# If all values are identical, the standard deviation is 0, leading to division by zero.
# StandardScaler handles this by setting all normalized values to 0.0.
df_edge['ConstantComp_Normalized_Zscore'] = scaler_zscore.fit_transform(df_edge[['ConstantComp']])

print("\nDataFrame with constant 'ConvertedCompYearly' (edge case for Z-score):")
print(df_edge)
print(f"Mean of 'ConstantComp_Normalized_Zscore': {df_edge['ConstantComp_Normalized_Zscore'].mean():.2f}")
print(f"Standard Deviation of 'ConstantComp_Normalized_Zscore': {df_edge['ConstantComp_Normalized_Zscore'].std():.2f}")
```

### Section 4: Visualization of Normalized Data
##### Task 7: Visualize the distribution of `ConvertedCompYearly`, `ConvertedCompYearly_Normalized`, and `ConvertedCompYearly_Zscore`

Visualization helps you understand how normalization changes the data distribution. In this task, create histograms for the original ConvertedCompYearly, as well as its normalized versions (ConvertedCompYearly_MinMax and ConvertedCompYearly_Zscore). This will help you compare how each normalization technique affects the data range and distribution.


```python
## Write your code here
import matplotlib.pyplot as plt # Added for plotting
import seaborn as sns # Added for enhanced plotting
plt.figure(figsize=(15, 5)) # Set up the figure size for better readability

# Plot original distribution
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st plot
sns.histplot(df['ConvertedCompYearly'], kde=True, bins=5, color='skyblue')
plt.title('Original ConvertedCompYearly')
plt.xlabel('Compensation')
plt.ylabel('Frequency')

# Plot Min-Max Normalized distribution
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd plot
sns.histplot(df['ConvertedCompYearly_Normalized_MinMax'], kde=True, bins=5, color='lightcoral')
plt.title('Min-Max Normalized')
plt.xlabel('Normalized Compensation')
plt.ylabel('Frequency')

# Plot Z-score Normalized distribution
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd plot
sns.histplot(df['ConvertedCompYearly_Normalized_Zscore'], kde=True, bins=5, color='lightgreen')
plt.title('Z-score Normalized')
plt.xlabel('Standardized Compensation')
plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show() # Display the plots
```
